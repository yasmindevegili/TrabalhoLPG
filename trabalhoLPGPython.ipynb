---



````python
notas = pd.read_csv("/content/tblnotas.csv", sep=";", encoding="latin-1")
display(notas)
````



---


`Função para separar todas as notas do mesmo aluno em colunas`


---



````python
def separa_notas(df_alunos, df_notas):
  nota1 = df_notas.query("id_nota == 1")['nota']
  nota2 = df_notas.query("id_nota == 2")['nota']
  nota3 = df_notas.query("id_nota == 3")['nota']
  notas = {
      'Nota 1' : nota1.values,
      'Nota 2' : nota2.values,
      'Nota 3' : nota3.values
  }
  tabela_trimestres = pd.DataFrame(notas)
  media = round(tabela_trimestres.mean(axis = 1), 2)
  tabela_trimestres['Média'] = media
  tabela_geral = df_alunos.join(tabela_trimestres)
  return tabela_geral
````



---


`Aplicando e printando a função`


---



````python
tabela_final = separa_notas(alunos, notas)
tabela_final
````



---


`Função para gerar arquivo a partir da tabela`


---



````python
def gravar_tabela(df, string=None):
  csv = df.to_csv(string, index=False, encoding="latin-1", sep=";")
  return csv

tabela_gravada = gravar_tabela(tabela_final)
print(tabela_gravada)
````



---


`Função para consultar a tabela`


---



````python
def consulta_tabela(df, id = None, nome = None):
  if id is not None:
    content = df.query(f'id == {id}')
    return content
  elif nome is not None:
    content = df.query(f'Nome == "{nome}"')
    return content

student_content = consulta_tabela(tabela_final, 1033253163, "YASMIN DEVEGILI")
display(student_content)
````



---


`Função para excluir um registro da tabela`


---



````python
def excluir_registro(df, id):
  df.loc[df.id == id, 'Excluido'] = 1
  return df

tabela_final = excluir_registro(tabela_final, 1033221237)
tabela_final.head()
````
